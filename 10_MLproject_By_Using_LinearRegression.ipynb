1. Analyzing the Impact of Temperature on Ice Cream Sales


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Download and prepare the data
data = pd.read_csv("https://raw.githubusercontent.com/NallamaruAdithyavardhan/Adithya-vardhan/refs/heads/main/README.md")
X = data[["Temperature (°C)"]].values
y = data[["Ice Cream Sales ($)"]].values
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#  visualize the data
plt.scatter(data["Temperature (°C)"], data["Ice Cream Sales ($)"])
plt.title("Temperature vs Ice Cream Sales")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales ($)")
plt.grid(True)
plt.show()
```

```python

# Perform linear regression
model = LinearRegression()
model.fit(X, y)
```

```python
# Make a prediction for Ice Cream Sales
X_new=[[15]]
print(model.predict(X_new))
```

```python
data['Predicted Ice Cream Sales'] = model.predict(X)
```

```python
# Plotting the regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature (°C)', y='Ice Cream Sales ($)', data=data, label='Actual Sales')
plt.plot(data['Temperature (°C)'], data['Predicted Ice Cream Sales'], color='red', label='Regression Line')
plt.title('Temperature vs Ice Cream Sales with Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales ($)')
plt.legend()
plt.grid(True)
plt.show()
```
